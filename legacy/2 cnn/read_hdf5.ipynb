```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import time
import cv2
```

```python
def hdf5_summary(file, label):
    hdf = h5py.File(file, 'r')
    ls = list(hdf.keys())
    print(f"summary of {label}")
    print(f"datasets:", ls)

    sum_imgs = 0
    for key in ls:
        sum_imgs += hdf[key].shape[0]
    print("total number of images:", sum_imgs)
```
